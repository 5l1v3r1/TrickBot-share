```python
# SHA

import pefile
import base64

key = '5Cjd3K0ifHTXRPUIbtyV/zA9lmG7E42WB61scJSQpkqge+Dax8MvhLZnowNruOYF'
std_b64 = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/"


pe = pefile.PE("PATH_TO_TRICKBOT_LOADER_BINARY")
encoded_blob = pe.sections[1].get_data()[976:6865]
data = encoded_blob.decode("utf-8") 

for s in data.split('\x00'):
    s = s.translate(str.maketrans(key,std_b64))
    if len(s)%4 != 0:
        s += '='*(4 - len(s)%4)
    print(base64.b64decode(s))
```
